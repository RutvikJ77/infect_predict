Data visualisation and EDA analysis


```python
import pandas as pd
import matplotlib.pyplot as plt
```

Basic file functions and analysis

```python
ob1=pd.read_csv('/content/observations_1.csv')
ob2=pd.read_csv('/content/observations_2.csv')
ob3=pd.read_csv('/content/observations_3.csv')
```

```python
total1=[ob1['infected_unvaccinated'].sum(),ob1['infected_vaccinated'].sum()]
labels1=['unvaccinated & infected','vaccinated & infected']
explode=(0,0.2)
fig,ax=plt.subplots()
ax.pie(total1,autopct='%1.0f%%',explode=explode)
ax.set_title('comparison of vaccinated and unvaccinated people gettig infected of observation 1')
plt.legend(labels1,loc=3)
```

```python
total2=[ob2['infected_unvaccinated'].sum(),ob2['infected_vaccinated'].sum()]
labels2=['unvaccinated & infected','vaccinated & infected']
explode=(0.2,0)
fig,ax=plt.subplots()
ax.pie(total2,autopct='%1.0f%%',explode=explode)
ax.set_title('comparison of vaccinated and unvaccinated people gettig infected of observation 2')
plt.legend(labels2,loc=3)
```

```python
total3=[ob1['infected_unvaccinated'].sum(),ob1['infected_vaccinated'].sum()]
labels3=['unvaccinated & infected','vaccinated & infected']
explode=(0.2,0)
fig,ax=plt.subplots()
ax.pie(total3,autopct='%1.0f%%',explode=explode)
ax.set_title('comparison of vaccinated and unvaccinated people gettig infected of observation 3')
plt.legend(labels3,loc=3)
```

```python
total=[total1[0]+total2[0]+total3[0], total1[1]+total2[1]+total3[1]]
labels=['unvaccinated & infected','vaccinated & infected']
explode=(0.2,0)
fig,ax=plt.subplots()
ax.pie(total,autopct='%1.0f%%',explode=explode)
ax.set_title('comparison of vaccinated and unvaccinated people getting infected combining all the observations')
plt.legend(labels,loc=3)
```

```python
ob1=(ob1-ob1.mean())/ob1.std()
ob2=(ob2-ob2.mean())/ob2.std()
ob3=(ob3-ob3.mean())/ob3.std()
```

```python
ob1['ob2_unvax'] = ob2['infected_unvaccinated']
ob1['ob3_unvax'] = ob3['infected_unvaccinated']
ob1['ob2_vax'] = ob2['infected_vaccinated']
ob1['ob3_vax'] = ob3['infected_vaccinated']
ob1=(ob1-ob1.mean())/ob1.std()
```
